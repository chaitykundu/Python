# y=x+20sin(x/10

```python
x_data = np.arange(100,step=0.1)
y_data = x_data +20*np.sin(x_data/10)
mlt.plot(x_data,y_data)
mlt.xlabel('x_axis',fontsize=20,color='green')
mlt.ylabel('y_axis' , fontsize=20,color='green')
mlt.show()
```

```python
#Pie Chart
activities = ['reading', 'sleeping', 'walking','working']
slice = [5,8,2,9]
colr = ['blue', 'magenta', 'green', 'red']
explode = (0,0,0,0)
plt.pie(slice, labels = activities, colors= colr, explode = explode, startangle = 120, shadow= True, autopct= '%0.2f%%')
plt.show()
```
