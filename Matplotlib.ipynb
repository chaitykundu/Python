### Matplotlib: Matplotlib is a low level graph plotting library in python that serves as a visualization utility.Most of the Matplotlib utilities lies under the pyplot submodule.

```python
#Draw a line in a diagram from position (0,0) to position (6,250):
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([1, 6])
ypoints = np.array([4, 250])
plt.plot(xpoints, ypoints)
plt.plot(xpoints, ypoints,'*')
plt.show()
```

### If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

```python
import matplotlib.pyplot as plt
import numpy as np
yplots = np.array([2,4,6,0,10,13])
#plt.plot(yplots, marker= 'o')
plt.plot(yplots,'o:b', ms=10, mec='r', mfc='b') #represents format, color ms=marker size, mec=markeredgecolor
plt.show()
```

```python
#linestyle
ypoints = np.array([3,6,5, 8,9])
xpoints = np.array([5, 2,6,9,10])
plt.plot(ypoints, linestyle = '--', c="c", linewidth= '5.5') #dotted= :, dash dot= -.
plt.plot(xpoints,linestyle='-.',c="r", linewidth= '5.5')
plt.show()
```

```python
#labels
a = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
b = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(a, b)

plt.title("Sports Watch Data", loc='left',c='g') # by defalt loc = center
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.grid(linestyle='--', color='b')

plt.show()
```

```python
#The subplot() function takes three arguments that describes the layout of the figure.
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

x1 = np.array([0, 1, 2, 3])
y1 = np.array([10, 20, 30, 40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.title("SALES")
plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.title("INCOME")
plt.suptitle("My Shop")
plt.show()
```

```python
#viridis colormap
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
plt.scatter(x, y, s=sizes, alpha=0.5)
```

```python
#combinecolor size and alpha
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
